```python
import matplotlib.pyplot as plt
import json as json
import numpy as np
```

```python
with open('data/weight_link_counts.json') as fil:
    weights = json.load(fil)
```

```python
counts_0 = np.array([[float(row['weight_val']),int(row['n_weights'])] for row in weights if row['link_count'] == '0'])
counts_1 = np.array([[float(row['weight_val']),int(row['n_weights'])] for row in weights if row['link_count'] == '1'])
counts_2 = np.array([[float(row['weight_val']),int(row['n_weights'])] for row in weights if row['link_count'] == '2'])
counts_3 = np.array([[float(row['weight_val']),int(row['n_weights'])] for row in weights if row['link_count'] == '3'])
counts_4 = np.array([[float(row['weight_val']),int(row['n_weights'])] for row in weights if row['link_count'] == '4'])
counts_5 = np.array([[float(row['weight_val']),int(row['n_weights'])] for row in weights if row['link_count'] == '5'])
```

```python
plt.plot(counts_0[:,0],counts_0[:,1],color='C0',label='links = 0')
plt.legend()
plt.xlim([0,1])
plt.show()

plt.plot(counts_1[:,0],counts_1[:,1],color='C1',label='links = 1')
plt.legend()
plt.xlim([0,1])
plt.show()

plt.plot(counts_2[:,0],counts_2[:,1],color='C2',label='links = 2')
plt.legend()
plt.xlim([0,1])
plt.show()

plt.plot(counts_3[:,0],counts_3[:,1],color='C3',label='links = 3')
plt.legend()
plt.xlim([0,1])
plt.show()

plt.plot(counts_4[:,0],counts_4[:,1],color='C4',label='links = 4')
plt.legend()
plt.xlim([0,1])
plt.show()

plt.plot(counts_5[:,0],counts_5[:,1],color='C5',label='links = 5')
plt.legend()
plt.xlim([0,1])
plt.show()
```

```python
counts_0_norm = np.array([[r[0],r[1]/sum(counts_0[:,1])*1000] for r in counts_0])
counts_1_norm = np.array([[r[0],r[1]/sum(counts_1[:,1])*1000] for r in counts_1])
counts_2_norm = np.array([[r[0],r[1]/sum(counts_2[:,1])*1000] for r in counts_2])
counts_3_norm = np.array([[r[0],r[1]/sum(counts_3[:,1])*1000] for r in counts_3])
counts_4_norm = np.array([[r[0],r[1]/sum(counts_4[:,1])*1000] for r in counts_4])
counts_5_norm = np.array([[r[0],r[1]/sum(counts_5[:,1])*1000] for r in counts_5])
```

```python
plt.plot(counts_0_norm[:,0],counts_0_norm[:,1],color='C0',label='links = 0')
plt.legend()
plt.ylim([0,50])
plt.xlim([0,1])
plt.show()

plt.plot(counts_1_norm[:,0],counts_1_norm[:,1],color='C1',label='links = 1')
plt.legend()
plt.ylim([0,50])
plt.xlim([0,1])
plt.show()

plt.plot(counts_2_norm[:,0],counts_2_norm[:,1],color='C2',label='links = 2')
plt.legend()
plt.ylim([0,50])
plt.xlim([0,1])
plt.show()

plt.plot(counts_3_norm[:,0],counts_3_norm[:,1],color='C3',label='links = 3')
plt.legend()
plt.ylim([0,50])
plt.xlim([0,1])
plt.show()

plt.plot(counts_4_norm[:,0],counts_4_norm[:,1],color='C4',label='links = 4')
plt.legend()
plt.ylim([0,50])
plt.xlim([0,1])
plt.show()

plt.plot(counts_5_norm[:,0],counts_5_norm[:,1],color='C5',label='links = 5')
plt.legend()
plt.ylim([0,50])
plt.xlim([0,1])
plt.show()
```
